```python
import pandas as pd

df = pd.read_csv('CleaningV2.txt', header = 0)
df = pd.read_csv('CleaningV2.txt', names = ['#YY'  ,'MM' ,'DD' ,'hh' ,'mm' ,
'WDIR' ,'WSPD', 'GST' , 'WVHT',  'DPD' ,  'APD' , 'MWD' ,  'PRES' , 'ATMP' , 'WTMP' , 'DEWP' , 'VIS' , 'TIDE'])
# Imports the text file and sets up the column names
df = df.drop(df.index[0:2])
df=df.reset_index(drop=True)
#got rid of useless rows and reset the index back to 0

df= df.astype(float)
#changes all the data to float

df = df[(df.hh >= 6) & (df.hh <= 17)]
df=df.reset_index(drop=True)
#only grabs the hours between 6am - 6pm
df = df[df != 99]
df = df[df!= 999]
df = df.fillna(0)
# removing useless values and replaces them with 0

df = df.groupby(df.index // 36).mean()




df
```
